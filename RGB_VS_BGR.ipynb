```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

Loading an Image using open cv

```python
img = cv.imread('/content/rgb.jpeg')
plt.imshow(img)
```

##### RGB <-> BGR
       BGR <-> RGB

##2.CONVERTING BGR TO RGB

```python
img1 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img = cv.cvtColor(img, cv.COLOR_RGB2BGR)
plt.imshow(img)
```
